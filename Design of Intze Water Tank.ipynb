```python
import math
import pandas as pd
C = float(input('Enter the capacity of tank in m3: '))
D = float(input('Assume the diameter of tank in m: '))
Do = float(input('Assume the secondary diameter of tank in m (usually 0.6 D): '))
h1 = float(input('Assume the height of top dome in m (usually 0.2 D): '))
h2 = float(input('Assume the height of bottom dome in m (usually 0.2 D): '))
ho = float(input('Assume the height of conical dome in m: '))

fck = int(input('Enter the grade of concrete: '))
fy = int(input('Enter the design strength of steel: '))

print('From Table 1 of IS 3370 part 2, 2009, œÉct is chosen according to grade of concrete.')
map_table_1_3370P2 = {
        15  : (1.1 , 1.5  ),
        20  : (1.2 , 1.7  ),
        25  : (1.3 , 1.8  ),
        30  : (1.5 , 2.0  ),
        35  : (1.6 , 2.2  ),
        40  : (1.8 , 2.4  ),
        45  : (2.0 , 2.6  ),
        50  : (2.1 , 2.8  ),
        }
Grade = list(map_table_1_3370P2.keys())
if fck in Grade:
    œÉct1 = map_table_1_3370P2[fck]
    œÉct = œÉct1[0]
    print('The value of œÉct is=',œÉct,'N/mm2')
else:
    print('The grade of concrete is not present in Table 1 of IS 3370 (2009).')

print('From Table 4 of IS 3370 part 2, 2009, œÉst is chosen according to type of stress and type of steel.')
SteelType = str(input('Enter P for plain round mild steel bar and H for High Strength Deformed Bars.'))
StressCondition = str(input('Enter T for :Tensile Stress in Member under direct tension, bending and shear and C for compressive stress in column subjected to direct load.'))

if SteelType == 'P' and StressCondition == 'T':
    œÉst = 115
elif SteelType == 'H' and StressCondition == 'T':
    œÉst = 130
elif SteelType == 'P' and StressCondition == 'C':
    œÉst = 125
elif SteelType == 'H' and StressCondition == 'C':
    œÉst = 140
print('The value of œÉst is=',œÉst,'N/mm2')

map_table_2_3370P2 = {
        10  : (3    , 2.5 , 0   ),
        15  : (5    , 4   , 0.6 ),
        20  : (7    , 5   , 0.8 ),
        25  : (8.5  , 6   , 0.9 ),
        30  : (10   , 8   , 1   ),
        35  : (11.5 , 9   , 1.1 ),
        40  : (13   , 10  , 1.2 ),
        45  : (14.5 , 11  , 1.3 ),
        50  : (16   , 12  , 1.4 ),
        }
Grade = list(map_table_2_3370P2.keys())
if fck in Grade:
    œÉcbc1 = map_table_2_3370P2[fck]
    œÉcbc = œÉcbc1[0]
    œÉcc = œÉcbc1[1]
    print('The value of œÉcbc is=',œÉcbc,'N/mm2')
    print('The value of œÉcc is=',œÉcc,'N/mm2')
else:
    print('The grade of concrete is not present in Table 2 of IS 3370 (2009).')
```

```python
R1 = round(((Do/2)**2 + h2**2)/(2*h2),2)
print('The radius of top dome =',R1)
h = round((C + (3.14 / 3 * (h2**2) * (3 * R1 - h2 ) ) - ( 3.14 / 12 * ho * ( D**2 + Do**2 + D * Do ) )) / ( 3.14 / 4 * D**2 ),2)
print('The required value of h=',h)
h = int(input('Enter the final value of height including free board:'))
```

```python
#Design of Top Dome
TDThickness = int(input('Assume the thickness of top dome in mm (usually 75 - 100):'))
LL = float(input('Enter the live load in kN/m2: '))
w = TDThickness / 1000 * 25
print('The self weight of dome = ',w,' kN/m.')
TDTL = w + LL
print('The total load = ',TDTL,'kN/m.')
R2 = ((D / 2)**2 + h1**2) / (2 * h1)

sinŒ∏1 =  (D/2) / R2
x = math.degrees((math.asin(sinŒ∏1)))
cosŒ∏ = ((math.cos(math.radians(x))))
T1 = round(TDTL * R2 / (1 + cosŒ∏),2)
print('The meridional force on top dome, T1 =',T1,'kN/m.')
ST1 = round(T1 / 100,2)
print('The meridional stress on top dome, ST1 =',ST1,'N/mm2')
if ST1 < œÉcc:
    print('SAFE (< ',œÉcc,' N/mm2)')
else:
    print('UNSAFE (>= ',œÉcc,' N/mm2)')

T2 = round(TDTL * R2 * (cosŒ∏ - 1 / (1 + cosŒ∏)),2)
print('The hoop force, T2 =',T2,'kN/m.')
ST2 = round(T2 / TDThickness,2)
print('The hoop stresses, ST2 =',ST2,'N/mm2')
if ST1 < œÉcbc:
    print('SAFE (< ',œÉcbc,' N/mm2)')
else:
    print('UNSAFE (>=',œÉcbc,'  N/mm2)')
    
print('Since both forces do not exceed their stresses, we will only provide nominal reinforcement. (0.30 %)')
Ast = 0.30 / 100 * 1000 * TDThickness
print('The area of steel required =',Ast,'mm2')
print('For meridional forces, steel is provided radially.')
print('For hoop forces,steel is provided circumferentially.')
```

```python
#DESIGN OF TOP RING BEAM
print('The horizontal component of the Meridional Force will act on the top ring beam.')
W1 = round(T1 * cosŒ∏,2)
print('The horizontal component of Meridional Force=',W1,'kN/m.')
THT = round(W1 * D / 2,2)
print('The total hoop tension in beam =',THT,'kN.')
TRAstreq = round(THT * 1000 / œÉst,2)
print('The area of steel required=',TRAstreq,'mm2.')
TRAstprov = float(input('Enter the area of steel provided: '))

b = int(input('Enter the breadth of the beam:'))
m = 280 / (3 * œÉcbc)
BeamD = round(( THT * 1000 / œÉct - ( m - 1 ) * TRAstreq ) / b,2)
print('The value of D should be greater than equals to =',BeamD,'mm.')
BeamD = int(input('Enter the overall depth in mm:'))
```

```python
import numpy as np
#Design of Cylindrical Wall
TableData = [('Height from top (m)','Tension (kN)',' Area of Steel Required (mm2)','Area in each face (mm2)')]
for i in range(1,h+1):
    T = 10 * i * D / 2
    AST = round(60 * i * 1000 /œÉst ,3)
    AST2 = AST / 2
    TableData.append((i,T,AST,AST2))
df = pd.DataFrame(data=TableData)
df
```

```python
List1 = TableData[h]
MaxTension= List1[1]
MaxAst = float(input('Enter the Maximum Ast provided:'))
CWallT = round(( T * 1000 / œÉct - MaxAst ) / 1000,2)
print('The thickness of cylindrical wall comes out to be',CWallT,'mm.')
CWallT = int(input('Enter the final thickness of cylindrical wall.'))

print('We provide minimum reinforcement steel (0.30%).')
CWAstmin = 0.30 / 100 * 1000 * CWallT
print('The minimum area of distribution steel =',CWAstmin,'mm2')
CWAstminEach = CWAstmin / 2
print('The area of distribution steel in each face =',CWAstminEach,'mm2')
CWAstminprov = float(input('Enter the area of distribution steel provided in each face.'))
```

```python
#Design of Bottom Ring Beam

print('The loading in the bottom ring beam comes from the components above it and the water pressure.')

sinŒ∏ = ((math.sin(math.radians(x))))

LTD = round(T1 * sinŒ∏,2)
print('The load per meter from top dome =',LTD,'kN/m.')

LRB = b / 1000 * BeamD / 1000 * 25
print('The load per meter from top ring beam =',LRB,'kN/m.')

BRBDepth = int(input('Assume the depth of bottom ring beam:'))
BRBWidth = int(input('Assume the width of bottom ring beam:'))
h11 = h - BeamD / 1000 - BRBDepth / 1000
print('The effective height of wall=',h11,'m.')

LW = round(CWallT / 1000 * h11 * 25 ,2)
print('The load due to wall =',LW,'kN/m.')

LSW = BRBDepth / 1000 * 25 * BRBWidth / 1000
print('The self weight of Bottom Ring Beam =',LSW,'kN/m.')

TLBRB = round(LSW + LW + LRB + LTD,2)
print('The total vertical loading in Bottom Ring Beam =',TLBRB,'kN/m.')

Œ≤ = int(input('Enter the angle made by the slanting slab connecting bottom ring beam with spherical dome beam with respect to vertical.'))
HF1 = round(TLBRB * math.tan(math.radians(Œ≤)),2)
print('The horizontal force acting on the beam due to the vertical loads=',HF1,'kN/m.')
HF2 = round(10 * ( h - BRBDepth / 2 / 1000 ) * BRBDepth / 1000,2)
print('The horizontal force acting on the beam due to the water pressure=',HF2,'kN/m.')
HF = HF1 + HF2
print('The total horizontal forces acting on the beam due to the water pressure and vertical loads =',HF,'kN/m.')

BRBHT = round(HF * D / 2,2)
print('The hoop tension in bottom ring beam =',BRBHT,'kN.')
BRBAst = round(BRBHT * 1000 / œÉst,2)
print('The area of steel required in bottom ring beam =',BRBAst,'mm2.')
BRBAstprov = float(input('Enter the area of steel provided:'))

BRBœÉct = round(BRBHT * 1000 / ((BRBWidth * BRBDepth) + (m - 1) * BRBAstprov),2)
print('The value of œÉct from calcuation =',BRBœÉct,'N/mm2.')

if BRBœÉct < œÉct:
    print('SAFE.')
else:
    print('UNSAFE.')
```

```python
#Design of Conical Dome:
CDAvgDia = ( D + Do ) / 2
print('The average diameter of dome =',CDAvgDia,'m.')
CDAvgDepth = (h + ( h + ho ) ) / 2
print('The average water depth of dome =',CDAvgDepth,'m.')
WWater = round(3.14 * CDAvgDia * CDAvgDepth * ( ( D - Do ) / 2 ) * 10,2)
print('The weight of water supported by conical dome =',WWater,'kN.')
CDSlabWidth = round( ( ho ** 2 + ( ( D - Do ) / 2 ) ** 2 ) ** 0.5, 2)
print('The width of slab =',CDSlabWidth,'m.')

CDSlabThickness = int(input('Assume the thickness of the dome:'))

CDSlabWeight = round(3.14 * CDAvgDia * CDSlabThickness / 1000 * CDSlabWidth * 25,2)
print('The total self weight of conical dome slab =',CDSlabWeight,'kN.')

UpperCDLoads = round(3.14 * D * TLBRB,2)
print('The loads from the upper components on conical dome =',UpperCDLoads,'kN.')

CDTL = round(UpperCDLoads + CDSlabWeight + WWater,2)
print('The total loads acting on the conical dome =',CDTL,'kN.')

CDTLPM = round(CDTL / ( 3.14 * Do ),2)
print('The total loads acting on the conical dome per meter =',CDTLPM,'kN/m.')

T3 = round(CDTLPM / math.cos(math.radians(Œ≤)),2)
print('The meridional force in conical dome, T3=',T3,'kN.')

ST3 = round(T3 / CDSlabThickness,2)
print('The meridional stress in conical dome =',ST3,'N/mm2.')

if ST3 < œÉcc:
    print('SAFE ( <',œÉcc,'N/mm2 )')
else:
    print('UNSAFE ( >=',œÉcc,'N/mm2 )')

    
print('Now, we will calculate hoop tension at the top, bottom and center of the dome.')
q = round(CDSlabThickness / 1000 * 25,2)
print('The self weight of slab, q=',q,'kN/m2.')

TopHT1 = 10 * h * (1 / math.sin(math.radians(90 - Œ≤)))  * D / 2
TopHT2 = q * ( 1 / math.tan(math.radians(90 - Œ≤))) * D / 2
TopHT = round(TopHT1 + TopHT2,2)
print('The hoop tension at the top of conical dome =',TopHT,'kN.')

MidHT1 = 10 * ( ( h + ( h + ho ) ) / 2 )  * ( 1 / (math.sin( math.radians( 90 - Œ≤ ) ) ) ) * CDAvgDia / 2
MidHT2 = q * ( 1 / ( math.tan( math.radians( 90 - Œ≤) ) ) ) * CDAvgDia / 2
MidHT = round(MidHT1 + MidHT2,2)
print('The hoop tension at the middle of conical dome =',MidHT,'kN.')

BottomHT1 = 10 * (h + ho) * (1/(math.sin(math.radians(90 - Œ≤) ) ) )  * Do / 2
BottomHT2 = q * ( 1 / ( math.tan( math.radians( 90 - Œ≤) ) ) ) * Do / 2
BottomHT = round(BottomHT1 + BottomHT2,2)
print('The hoop tension at the bottom of conical dome =',BottomHT,'kN.')

MaxCDHT = max(TopHT,MidHT,BottomHT)
print('The maximum tension from 3 locations =',MaxCDHT,'kN.')

CDAst = round(MaxCDHT * 1000 / œÉst,2)
print('The area of steel required =',CDAst,'mm2')

CDAstEach = CDAst / 2
print('The area of steel in each face =',CDAstEach,'mm2')

CDAstprov = float(input('Enter the area of steel provided in each face in mm2:'))

print('Now, we provide minimum distribution steel (0.30%).')
CDAstmin = 0.30 / 100 * 1000 *  CDSlabThickness
print('The minimum area of distribution steel required =',CDAstmin,'mm2')
CDAstminEach = CDAstmin / 2
print('The minimum area of distribution steel required for each face =',CDAstminEach,'mm2')

CDAstminEachProv = float(input('Enter the area of steel provided for each face:'))

CDœÉct = round(MaxCDHT * 1000 / ( ( 1000 * CDSlabThickness) + (m - 1) * 2 * CDAstminEachProv ),2)
print('The calcualted value of œÉct =',CDœÉct,'N/mm2.')

if ST3 < œÉct:
    print('SAFE')
else:
    print('UNSAFE')
```

```python
#Design of Bottom Spherical Dome:
print('Only meridional force acts on the bottom spherical dome.')
BSDThickness = int(input('Assume the thickness of bottom sheprical dome in mm:'))
BSDSW = BSDThickness / 1000 * 25
print('The self weight of bottom spherical dome =',BSDSW,'kN/m2')
BSDWL = ( h + ho - h2 ) * 10
print('The water load on bottom spherical dome =',BSDWL,'kN/m2')
BSDTL = BSDWL + BSDSW
print('The total loading on bottom spherical load =',BSDTL,'kN/m2')
R3 = round(( ( Do / 2 ) ** 2 + ( h2 ** 2 ) ) / ( 2 * h2 ),2)
print('The Radius of bottom shperical dome =',R3,'m.')
sinŒ∏2 = (Do / ( 2 * R3 ) )
Œ∏2 = math.asin(sinŒ∏2)
cosŒ∏2 = math.cos(Œ∏2)
T4 = round(BSDTL * R3 / ( 1 + cosŒ∏2 ),2)
print('The total meridional force on bottom spherical dome =',T4,'kN/m.')
ST4 = round(T4 / BSDThickness,2)
print('The meridional stress on bottom spherical dome =',ST4,'N/mm2.')
if ST4 < œÉcc:
    print('SAFE ( < ',œÉcc,' N/mm2 )')
else:
    print('UNSAFE ( >=',œÉcc,' N/mm2 )')
T5 = round(BSDTL * R3 * ( cosŒ∏2 - 1 / ( 1 + cosŒ∏2 ) ),2)
print('The circumferential force on bottom spherical dome=',T5,'kN/m.')
ST5 = round(T5 / BSDThickness,2)
print('The circumferential stress on bottom spherical dome=',ST5,'N/mm2.')
if ST5 < œÉcc:
    print('SAFE ( < ',œÉcc,' N/mm2 )')
else:
    print('UNSAFE ( >=',œÉcc,' N/mm2 )')
print('We provide minimum reinforcement (0.30%).')
BSDAstmin = 0.30 / 100 * 1000 * BSDThickness
print('The minimum area of distribution steel required =',BSDAstmin,'mm2')
BSDAstminprov = float(input('Enter the area of distribution steel provided:'))
```

```python
#Design of Bottom Circular Beam:
print('We have inward thrust from conical dome and outward thrust from spherical dome acting horizontally on the bottom circular beam.')

ITCD = round(T3 * math.cos( math.radians(90 - Œ≤) ),2)
print('The inward thrust from conical dome =',ITCD,'kN/m, acting in +ve x-direction.')
OTSD = round(T4 * math.cos(Œ∏2),2)
print('The outward thrust from spherical dome =',OTSD,'kN/m, acting in -ve x-direction.')

if ITCD > OTSD:
    DIR = 'acting in +ve x-direction.'
else:
    DIR = 'acting in -ve x-direction.'

BCBNT = round(abs(ITCD - OTSD),2)
print('The net thrust on the bottom circular beam =',BCBNT,DIR)

BCBHoopForce = round(BCBNT * Do / 2,2)
print('The hoop force acting on the beam =',BCBHoopForce,'kN.')

BCBWidth = int(input('Assume the width of the beam:'))
BCBDepth = int(input('Assume the depth of the beam:'))
BCBCover = int(input('Assume the cover of the beam:'))
BCBEffDepth = BCBDepth - BCBCover
print('The effective depth of the bottom circular beam =',BCBEffDepth)
BCBHoopStress = round(BCBHoopForce * 1000 / ( BCBDepth * BCBWidth ),2)
print('The bottom circular beam hoop stress =',BCBHoopStress,'N/mm2.')
if BCBHoopStress < œÉcc:
    print('SAFE ( < ',œÉcc,' N/mm2 )')
else:
    print('UNSAFE ( >= ',œÉcc,' N/mm2 )')

VLCD = round(T3 * math.sin( math.radians(90 - Œ≤) ),2)
print('The vertical force from conical dome =',VLCD,'kN/m.')
VLSD = round(T4 * math.sin(Œ∏2),2)
print('The vertical force from spherical dome =',VLSD,'kN/m.')
BCBSW = round(BCBDepth / 1000 * BCBWidth / 1000 * 25,2)
print('The self weight of bottom circular beam =',BCBSW,'kN/m.')
BCBVL = round(VLCD + VLSD + BCBSW,2)
print('The total vertical loading per meter on bottom circular beam =',BCBVL,'kN/m.')
BCBTVL = round(3.14 * Do *  BCBVL,2)
print('The total vertical loading on bottom circular beam =',BCBTVL,'kN.')
```

```python
map_table_Coeff4Supports = {
        4  : (90 , 0.0342 , 0.0176, 0.0053, 19.2),
        5  : (60 , 0.0148 , 0.0075, 0.0015, 12.733),
        8  : (45 , 0.0083 , 0.0041, 0.0006, 9.55 ),
        10 : (36 , 0.0054 , 0.0023, 0.0003, 7.50 ),
        12 : (30 , 0.0037 , 0.0014, 0.0017, 7.25 ),
        }
NoOfColumns = int(input('Enter the number of supports (4,5,8,10,12):'))
CoeffList = map_table_Coeff4Supports[NoOfColumns]
Angle = 3.14 / 180 * CoeffList[0]
C1 = CoeffList[1]
C2 = CoeffList[2]
C3 = CoeffList[3]
C4 = CoeffList[4]
MaxNegBM = round(C1 * BCBTVL *  Do / 2 ,2)
print('The maximum negative bending moment =',MaxNegBM,'kNm.')
MaxPosBM = round(C2 * BCBTVL *  Do / 2 ,2)
print('The maximum positive bending moment =',MaxPosBM,'kNm.')
MaxTorBM = round(C3 * BCBTVL *  Do / 2 ,2)
print('The maximum torsional bending moment =',MaxTorBM,'kNm.')

SupportSF = round(BCBVL * Do / 2  / 2 * 3.14 * 360 / 180 / NoOfColumns,2)
print('The shear force at support =',SupportSF)
Mtorsion = round(MaxTorBM * ( ( 1 + BCBDepth / BCBWidth ) / 1.7 ),2)
print('The value of Mt according to B-6.4.2 IS 456, 2000 =',Mtorsion)
```

```python
print('DESIGN OF SUPPORT SECTION:')
TotalNegBM = MaxNegBM + Mtorsion
print('The total negative bending moment at support =',TotalNegBM)
k = round(m * œÉcbc / ( m * œÉcbc + œÉst),2)
j = round(1 - k / 3,2)
Q = round(0.5 * œÉcbc * k * j,2)
BCBDepthCalc = round( ( TotalNegBM * ( 10**6 ) / ( Q * BCBWidth ) )**0.5, 2 )
print('The calculated depth of bottom circular beam =',BCBDepthCalc)
if BCBDepthCalc < BCBEffDepth:
    print('SAFE')
else:
    print('Depth of the bottom circular beam must be revised.')
    BCBEffDepth = int(input('Enter the new revised effective depth of Bottom circular beam:'))
BCBAst = round(TotalNegBM * (10 ** 6) / œÉst / j / BCBEffDepth,2)
print('The area of steel required for Bottom Circular Beam =',BCBAst)
MinAst4Section1 = round(0.85 / fy * BCBEffDepth * BCBWidth,2)
print('The minimum area of steel as per IS 456:2000 =',MinAst4Section1)
MinAst4Section2 = round(0.24 * fck ** 0.5 / fy * BCBWidth * BCBEffDepth,2)
print('The minimum area of steel as per IS 13920:2016 =',MinAst4Section2)
MinAst4Section = round(max(MinAst4Section1,MinAst4Section2),2)
print('The minimum area of steel required for bottom circular girder from IS 456 and IS 13920=',MinAst4Section)
BCBAstprov = float(input('Enter the area of steel provided:'))
```
